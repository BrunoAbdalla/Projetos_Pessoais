```python
#PROGRAMA QUE ANALISA OS ÚLTIMOS RESULTADOS DA MEGA-SENA

# Serão avaliados os resultados de 3 formas:

# - Modo genérico, com base apenas na parte de cima e de baixo do cartão;
# - Método dos quadrantes 2x2, onde o cartão é dividido em 4 quadrantes, conforme imagem 1;
# - Método dos quadrantes 2x3, onde o cartão é dividido em 6 quadrantes, conforme a imagem 2.
```

```python
#Importando a função para organizar a listagem dos números sorteados 
import operator

#Contadores de controle para organização e os de análise de resultados
cont = 0 
cont30 = cont60 = conte = contd = cont11 = cont21 = cont31 = cont41 = cont12 = cont22 = cont32 = cont42 = cont52= cont62 = 0
```

```python
#Definindo o lado esquerdo e direito do cartão
esquerdo = [1, 2, 3, 4, 5, 11, 12, 13, 14, 15, 21, 22, 23, 24, 25, 31, 32, 33, 34, 35, 41, 42, 43, 44, 45, 51, 52, 53, 54, 55]
direito = [6, 7, 8, 9, 10, 16, 17, 18, 19, 20, 26, 27, 28, 29, 30, 36, 37, 38, 39, 40, 46, 47, 48, 49, 50, 56, 57, 58, 59, 60]

#Difinindo os quadrantes do cartão (Modelo 2x2):
Q1_1 = [1, 2, 3, 4, 5, 11, 12, 13, 14, 15, 21, 22, 23, 24, 25]
Q2_1 = [6, 7, 8, 9, 10, 16, 17, 18, 19, 20, 26, 27, 28, 29, 30]
Q3_1 = [31, 32, 33, 34, 35, 41, 42, 43, 44, 45, 51, 52, 53, 54, 55]
Q4_1 = [36, 37, 38, 39, 40, 46, 47, 48, 49, 50, 56, 57, 58, 59, 60]

#Definindo os quadrantes do cartão (Modelo 2X3):
Q1_2 = [1, 2, 3, 11, 12, 13, 21, 22, 23]
Q2_2 = [4, 5, 6, 7, 14, 15, 16, 17, 24, 25, 26, 27]
Q3_2 = [8, 9, 10, 18, 19, 20, 28, 29, 30]
Q4_2 = [31, 32, 33, 41, 42, 43, 51, 52, 53]
Q5_2 = [34, 35, 36, 37, 44, 45, 46, 47, 54, 55, 56, 57]
Q6_2 = [38, 39, 40, 48, 49, 50, 58, 59, 60]
```

```python
#Resultados dos últimos concursos
Res1 = [11, 14, 36, 41, 54, 59]
Res2 = [20, 34, 38, 40, 49, 54]
Res3 = [1, 32, 35, 44, 45, 57]
Res4 = [1, 8, 21, 27, 36, 37]
Res5 = [4, 6 , 9, 31, 50, 56]
Res6 = [10, 15, 17, 20, 21, 35]
Res7 = [2, 17, 23, 28, 39, 46]
Res8 = [20, 33, 37, 38, 49, 50]
Res9 = [2, 7, 32, 46, 49, 55]
Res10 = [1, 40, 44, 45, 58, 60]
Res11 = [22, 30, 38, 39, 49, 56]
Res12 = [15, 18, 28, 42, 55, 60]
```

```python
#Lista com os últimos resultados
Todos = [Res1, Res2, Res3, Res4, Res5, Res6, Res7, Res8, Res9, Res10, Res11, Res12]

#Dicionário onde serão armazenados os números e suas ocorrências
Avaliação = {}

#Analisando os resultados dos sorteios e armazenando as ocorrências no dicionário 
for n in range(0, len(Todos)):
    for r in Todos[n]:
            if r in Avaliação.keys():
                Avaliação[r] = Avaliação[r] + 1
            else:
                Avaliação[r] = 1
                
#Analisando quais números saíram da parte de baixo e de cima do cartão:
            if 1 < r < 31:
                cont30 += 1
            else:
                cont60 += 1
                
#Analisando quais números saíram da parte esquerda ou direta do cartão
            if r in esquerdo:
                conte += 1
            else:
                contd += 1
                
#Analisando em quais quadrantes do cartão os números saíram no modelo 2x2  
            if r in Q1_1:
                cont11 += 1
            elif r in Q2_1:
                cont21 += 1
            elif r in Q3_1:
                cont31 += 1
            else:
                cont41 += 1
                
#Analisando em quais quadrantes do cartão os números saíram no modelo 2x3  
            if r in Q1_2:
                cont12 += 1
            elif r in Q2_2:
                cont22 += 1
            elif r in Q3_2:
                cont32 += 1
            elif r in Q4_2:
                cont42 += 1
            elif r in Q5_2:
                cont52 += 1
            else: 
                cont62 += 1

#Organizando os dados do dicionário para melhorar a resposta ao usuário
Organizado = sorted(Avaliação.items(), key=operator.itemgetter(0))
```

```python
#Retorno ao usuário
print('-=' * 34)
print(f'{"AVALIAÇÃO DOS ÚLTIMOS RESULTADOS":^68}')
print('-=' * 34)

#Apresentação do resultado geral
for k, v in Organizado:
        print(f'{k:^4} saiu {v:^4}', end=' | ')
        cont += 1
        if cont == 4:
            print()
            cont = 0
        else:
            pass       

print('-=' * 34)
print(f'* Nos sorteios tivemos {cont30} números da metade de cima do cartão.')
print(f'* Nos sorteios tivemos {cont60} números da metade de baixo do cartão.')
print(f'* Nos sorteios tivemos {conte} números da metade esquerda do cartão.')
print(f'* Nos sorteios tivemos {contd} números da metade de direita do cartão.')
print()
print('-=' * 34)
print(f'{"AVALIAÇÃO POR QUADRANTE (MODELO 2x2)":^68}')
print('-=' * 34)
print(f'* Nos sorteios tivemos {cont11} números no 1º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont21} números no 2º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont31} números no 3º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont41} números no 4º quadrante do cartão.')
print()
print('-=' * 34)
print(f'{"AVALIAÇÃO POR QUADRANTE (MODELO 2x3)":^68}')
print('-=' * 34)
print(f'* Nos sorteios tivemos {cont12} números no 1º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont22} números no 2º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont32} números no 3º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont42} números no 4º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont52} números no 5º quadrante do cartão.')
print(f'* Nos sorteios tivemos {cont62} números no 6º quadrante do cartão.')
```

```python
#SORTEADOR DE JOGOS DA MEGASENA BASEADO NO MÉTODO 2x2 COM AS ESTATÍSTICAS DO PROGRAMA 1

import numpy

#Definição dos quadrantes do cartão (Modelo 2x2):
Q1_1 = [1, 2, 3, 4, 5, 11, 12, 13, 14, 15, 21, 22, 23, 24, 25]
Q2_1 = [6, 7, 8, 9, 10, 16, 17, 18, 19, 20, 26, 27, 28, 29, 30]
Q3_1 = [31, 32, 33, 34, 35, 41, 42, 43, 44, 45, 51, 52, 53, 54, 55]
Q4_1 = [36, 37, 38, 39, 40, 46, 47, 48, 49, 50, 56, 57, 58, 59, 60]

#Ocorrência em cada quadrante:
OQ1 = 3656
OQ2 = 3747
OQ3 = 3841
OQ4 = 3678

#Tamanho total das amostras
Total = OQ1 + OQ2 + OQ3 + OQ4

#Percentual da ocorrência em cada quadrante em relação ao total
PQ1 = OQ1 / Total
PQ2 = OQ2 / Total
PQ3 = OQ3 / Total
PQ4 = OQ4 / Total

#Quantidade de números que serão sorteados em cada um dos quadrantes
n1 = int((PQ1 * 100) // 1)
n2 = int((PQ2 * 100) // 1)
n3 = int((PQ3 * 100) // 1)
n4 = int(100 - n1 - n2 - n3)
 
#Sorteando os números em cada um dos quadrantes:
SQ1 = numpy.random.choice(Q1_1, n1)
SQ2 = numpy.random.choice(Q2_1, n2)
SQ3 = numpy.random.choice(Q3_1, n3)
SQ4 = numpy.random.choice(Q4_1, n4)

#Quantidade de números sorteados em cada um dos quadrantes:
q1 = PQ1 * 6
q2 = PQ2 * 6
q3 = PQ3 * 6
q4 = PQ4 * 6
somaq = q1 + q2 + q3 + q4

d1 = (q1) - int(q1)
d2 = (q2) - int(q2)
d3 = (q3) - int(q3)
d4 = (q4) - int(q4)

decimais = (d1, d2, d3, d4)

if max(decimais) == d1:
    q1 = int(q1) + 1
    q2 = int(q2)
    q3 = int(q3)
    q4 = int(q4)
elif max(decimais) == d2:
    q1 = int(q1)
    q2 = int(q2) + 1
    q3 = int(q3)
    q4 = int(q4)
elif max(decimais) == d3:
    q1 = int(q1)
    q2 = int(q2)
    q3 = int(q3) + 1
    q4 = int(q4)
else:
    q1 = int(q1)
    q2 = int(q2)
    q3 = int(q3)
    q4 = int(q4) + 1

print(f'Controle de proporção: Serão sorteados: {q1} | {q2} | {q3} | {q4}')

Jogo = []

SorteadosQ1 = numpy.random.choice(SQ1, q1, False)
Jogo.append(SorteadosQ1[0])
SorteadosQ2 = numpy.random.choice(SQ2, q2, False)
Jogo.append(SorteadosQ2[:])
SorteadosQ3 = numpy.random.choice(SQ3, q3, False)
Jogo.append(SorteadosQ3[:])
SorteadosQ4 = numpy.random.choice(SQ4, q4, False)
Jogo.append(SorteadosQ4[:])

print(Jogo)
```

```python
#SORTEADOR DE JOGOS DA MEGASENA BASEADO NO MÉTODO 2x2 COM AS ESTATÍSTICAS DO PROGRAMA 1

import numpy

#-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#                AVALIAÇÃO POR QUADRANTE (MODELO 2x3)                
#-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=
#* Nos sorteios tivemos 11 números no 1º quadrante do cartão.
#* Nos sorteios tivemos 11 números no 2º quadrante do cartão.
#* Nos sorteios tivemos 13 números no 3º quadrante do cartão.
#* Nos sorteios tivemos 8 números no 4º quadrante do cartão.
#* Nos sorteios tivemos 22 números no 5º quadrante do cartão.
#* Nos sorteios tivemos 19 números no 6º quadrante do cartão.

#Definição dos quadrantes do cartão (Modelo 2x3):
Q1_2 = [1, 2, 3, 11, 12, 13, 21, 22, 23]
Q2_2 = [4, 5, 6, 7, 14, 15, 16, 17, 24, 25, 26, 27]
Q3_2 = [8, 9, 10, 18, 19, 20, 28, 29, 30]
Q4_2 = [31, 32, 33, 41, 42, 43, 51, 52, 53]
Q5_2 = [34, 35, 36, 37, 44, 45, 46, 47, 54, 55, 56, 57]
Q6_2 = [38, 39, 40, 48, 49, 50, 58, 59, 60]

#Ocorrência em cada quadrante:
OQ1 = 11
OQ2 = 11
OQ3 = 13
OQ4 = 8
OQ5 = 22
OQ6 = 19

#Tamanho total das amostras
Total = OQ1 + OQ2 + OQ3 + OQ4 + OQ5 + OQ6

#Percentual da ocorrência em cada quadrante em relação ao total
PQ1 = OQ1 / Total
PQ2 = OQ2 / Total
PQ3 = OQ3 / Total
PQ4 = OQ4 / Total
PQ5 = OQ5 / Total
PQ6 = OQ6 / Total

#Quantidade de números que serão sorteados em cada um dos quadrantes
n1 = int((PQ1 * 100) // 1)
n2 = int((PQ2 * 100) // 1)
n3 = int((PQ3 * 100) // 1)
n4 = int((PQ4 * 100) // 1)
n5 = int((PQ5 * 100) // 1)
n6 = 100 - n1 - n2 - n3 - n4 -n5
 
#Sorteando os números em cada um dos quadrantes:
SQ1 = numpy.random.choice(Q1_2, n1)
SQ2 = numpy.random.choice(Q2_2, n2)
SQ3 = numpy.random.choice(Q3_2, n3)
SQ4 = numpy.random.choice(Q4_2, n4)
SQ5 = numpy.random.choice(Q5_2, n5)
SQ6 = numpy.random.choice(Q6_2, n6)

#Quantidade de números sorteados em cada um dos quadrantes:
q1 = PQ1 * 6
q2 = PQ2 * 6
q3 = PQ3 * 6
q4 = PQ4 * 6
q5 = PQ5 * 6
q6 = PQ6 * 6

print(f'As proposções encontradas foram: Q1: {q1:.2f}, Q2: {q2:.2f}, Q3: {q3:.2f}, Q4: {q4:.2f}, Q5: {q5:.2f}, Q6: {q6:.2f}')
print('-' * 40)
print('Com base nessas proporções informe:')
q1 = int(input('Quantos números você quer que sejam sorteados no quadrante 1: '))
q2 = int(input('Quantos números você quer que sejam sorteados no quadrante 2: '))
q3 = int(input('Quantos números você quer que sejam sorteados no quadrante 3: '))
q4 = int(input('Quantos números você quer que sejam sorteados no quadrante 4: '))
q5 = int(input('Quantos números você quer que sejam sorteados no quadrante 5: '))
q6 = int(input('Quantos números você quer que sejam sorteados no quadrante 6: '))

Jogo = []

SorteadosQ1 = numpy.random.choice(SQ1, q1, False)
Jogo.append(SorteadosQ1[:])
SorteadosQ2 = numpy.random.choice(SQ2, q2, False)
Jogo.append(SorteadosQ2[:])
SorteadosQ3 = numpy.random.choice(SQ3, q3, False)
Jogo.append(SorteadosQ3[:])
SorteadosQ4 = numpy.random.choice(SQ4, q4, False)
Jogo.append(SorteadosQ4[:])
SorteadosQ5 = numpy.random.choice(SQ5, q5, False)
Jogo.append(SorteadosQ5[:])
SorteadosQ6 = numpy.random.choice(SQ6, q6, False)
Jogo.append(SorteadosQ6[:])

print('-'*40)
print(Jogo)
```
